```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="+",color="red",linewidth=2,label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
plt.xlim(0,5)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis,marker="D",color="Green", linewidth=2,label="Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fares($)")
ax.set_title("Pyber Fares by")
ax.legend("Boston")
ax.grid()
```

```python
plt.bar(x_axis,y_axis, color="Green",label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis,y_axis, color="magenta",label="Boston")
plt.gca().invert_yaxis()
plt.xlabel("Fares($)")
plt.ylabel("Dates")
plt.title("PyBer Fares by Month")
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis, color='cyan', label="Chicago")
ax.invert_yaxis()
ax.set_xlabel("Fares($)")
ax.set_ylabel("Date")
ax.legend()
```

```python
plt.scatter(y_axis,x_axis, c="red", label="Chicago")
plt.xlabel("Fare($)")
plt.xlim(0,50)
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
```
